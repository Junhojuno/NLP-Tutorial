### Bag of words model
- 단어를 토큰으로 만들어주고, 각 단어가 몇번등장 했는지 횟수를 vector로 만들어준다.
- n-gram을 사용하여, bigram, trigram등으로 몇개의 단어를 한 묶음으로 할지 정할 수 있음


### scikit-learn의 CountVectorizer를 통해 feature 생성
- 정규표현식을 사용해 토큰을 추출한다.
- 모두 소문자로 변환시키기 때문에 Good, good....등등 모두 같은 특성이 된다.
- 의미없는 특성을 많이 생성하기 때문에 적어도 두 개의 문서에 나타난 토큰만을 사용한다.
- min_df로 토큰이 나타날 최소 문서 개수를 지정할 수 있다.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
```

```python
vectorizer = CountVectorizer(analyzer='word',
                             tokenizer= None,
                             preprocessor = None,
                             stop_words = None,
                             min_df = 2,
                             ngram_range = (1, 3),
                             max_features = 20000)
vectorizer
```

```python
pipeline = Pipeline([('vect',vectorizer),
                    ])
```

```python
train = pd.read_csv("train_preprocessing.csv")
train.head()
```

```python
train_data_feature = pipeline.fit_transform(train.review_cleaning)
train_data_feature
```

```python
train_data_feature.shape
```

```python
vocab = vectorizer.get_feature_names() # 단어들이 뭐가 있는지 확인
print(len(vocab))
vocab[:10]
```

```python
# vector화된 feature를 확인해보자
dist = np.sum(train_data_feature, axis=0)

for tag, count in zip(vocab, dist):
    print(count, tag)
pd.DataFrame(dist, columns=vocab)
```
